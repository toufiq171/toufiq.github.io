## Script Analysis and Keyword Extraction

### Subtask:
Develop a component to analyze the input script, identify key themes, objects, actions, and emotional tones, and extract relevant keywords or phrases for video search.


**Reasoning**:
First, I will define a sample script. Then, I will import the `spaCy` library and load its small English language model. After processing the script with the loaded model, I will extract keywords and phrases by identifying nouns, verbs, and named entities. Finally, I will print the extracted keywords to demonstrate the analysis.



```python
import spacy

# 1. Define a sample input script
script = """The young detective, a brilliant mind with a penchant for cryptic crosswords, walked silently through the foggy London streets. A single raven cawed ominously from a gnarled oak tree. He was searching for clues related to the stolen 'Eye of Ra' diamond, a priceless artifact from the British Museum. The air was thick with tension and the faint scent of old parchment. Suddenly, a shadowy figure darted across an alley, carrying a large, heavy bag. The detective quickened his pace, his heart pounding with anticipation. This case was getting interesting. He knew he was close to unraveling the mystery.
"""

# 2. Import a natural language processing (NLP) library like `spaCy` and load an appropriate English language model
# You might need to download the model first if it's not already installed:
# !python -m spacy download en_core_web_sm
try:
    nlp = spacy.load("en_core_web_sm")
    print("spaCy model 'en_core_web_sm' loaded successfully.")
except OSError:
    print("Downloading spaCy model 'en_core_web_sm'. This may take a moment...")
    !python -m spacy download en_core_web_sm
    nlp = spacy.load("en_core_web_sm")
    print("spaCy model 'en_core_web_sm' downloaded and loaded successfully.")

# 3. Process the script using the loaded NLP model
doc = nlp(script)

# 4. From the processed script, extract a list of keywords and phrases
keywords = []

# Extract named entities
for ent in doc.ents:
    keywords.append(ent.text)

# Extract important nouns and verbs
for token in doc:
    # Consider proper nouns, common nouns, and verbs
    if token.pos_ in ("NOUN", "PROPN", "VERB") and not token.is_stop and not token.is_punct:
        keywords.append(token.lemma_)

# Remove duplicates and sort for cleaner output
keywords = sorted(list(set(keywords)))

# 5. Print the extracted keywords and phrases
print("\nExtracted Keywords and Phrases:")
print(keywords)
```

**Reasoning**:
Now that the user understands how to set up the Pexels API key, the next step is to install the `requests` library (if not already installed) to make HTTP requests to the Pexels API, retrieve the API key from Colab secrets, construct a search query using the extracted keywords, and then make the API call to search for videos.



```python
import requests
from google.colab import userdata

# 1. Retrieve API key securely
PEXELS_API_KEY = userdata.get("PEXELS_API_KEY")

if not PEXELS_API_KEY:
    print("Error: PEXELS_API_KEY not found in Colab Secrets. Please set it up as instructed.")
else:
    print("Pexels API Key successfully loaded.")

    # 2. Construct search query using keywords
    # For simplicity, we'll join the first few keywords to form a single query string.
    # In a more advanced scenario, you might process keywords more intelligently.
    search_query = " ".join(keywords[:5]) # Using the first 5 keywords
    print(f"\nSearch Query: {search_query}")

    # 3. Define Pexels API endpoint and headers
    PEXELS_VIDEO_SEARCH_URL = "https://api.pexels.com/videos/search"
    headers = {
        "Authorization": PEXELS_API_KEY
    }
    params = {
        "query": search_query,
        "per_page": 5, # Request 5 videos
        "orientation": "landscape"
    }

    # 4. Make the API call
    print("\nMaking API call to Pexels...")
    try:
        response = requests.get(PEXELS_VIDEO_SEARCH_URL, headers=headers, params=params)
        response.raise_for_status() # Raise an HTTPError for bad responses (4xx or 5xx)
        video_data = response.json()
        print("API call successful!")

        # 5. Parse the API response and extract relevant information
        found_videos = []
        if video_data and video_data.get("videos"):
            for video in video_data["videos"]:
                # Find the largest MP4 file
                highest_res_link = ""
                highest_res_width = 0
                for file in video.get("video_files", []):
                    if file.get("link") and file.get("file_type") == "video/mp4":
                        # Pexels API sometimes includes width/height in 'size' or directly
                        # Let's try to get width from 'width' or 'quality' if 'width' is missing
                        current_width = file.get("width")
                        if current_width and current_width > highest_res_width:
                            highest_res_width = current_width
                            highest_res_link = file["link"]

                found_videos.append({
                    "id": video.get("id"),
                    "url": video.get("url"),
                    "image_url": video.get("image"),
                    "duration": video.get("duration"), # Duration in seconds
                    "width": video.get("width"),
                    "height": video.get("height"),
                    "download_link": highest_res_link # Direct download link for the best quality MP4
                })
            print(f"Found {len(found_videos)} videos.")
            for i, vid in enumerate(found_videos):
                print(f"\nVideo {i+1}:")
                print(f"  ID: {vid['id']}")
                print(f"  Pexels URL: {vid['url']}")
                print(f"  Duration: {vid['duration']} seconds")
                print(f"  Resolution: {vid['width']}x{vid['height']}")
                print(f"  Download Link: {vid['download_link']}")
        else:
            print("No videos found for the given query.")

    except requests.exceptions.HTTPError as errh:
        print(f"Http Error: {errh}")
    except requests.exceptions.ConnectionError as errc:
        print(f"Error Connecting: {errc}")
    except requests.exceptions.Timeout as errt:
        print(f"Timeout Error: {errt}")
    except requests.exceptions.RequestException as err:
        print(f"Something went wrong: {err}")

# Store found_videos for subsequent steps
```
